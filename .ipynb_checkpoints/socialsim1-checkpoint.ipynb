## Primer ejercicio
Se plantea realizar una simulación simple en base al juego de jankenpón. Para la tarea se pide que añadir un código para revisar los resultados en vivo del torneo.

### 1) Se configuran las reglas, estrategias y los agentes

*Configuración de estrategias* como las opciones que los agentes pueden realizar en el juego.

```python
strategies=['Rock','Paper','Scissors']
```

*Configuración de reglas* como los indicadores de la relación entre entrategias (derrota-victoria)

```python
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0)}
```

*Configuración de agentes* que serán los que jueguen y tomen decisiones

```python
names=['Chayanne','Luis Miguel','El Puma','Maradona']
society=[{'name':n,'score':0,'strategy':None} for n in names]
society
```

*Configuración de toma de decisión* para los agentes y estrategias

```python
from random import choice

choice(strategies);
```

```python
import itertools

# pair is a tuple of dicts
for pair in itertools.combinations(society,2):
    print(pair)
```

```python
import itertools

# each dict
for player1,player2 in itertools.combinations(society,2):
    print(player1,player2)
```

### 2) Creación de la simulación

```python
society=[{'name':n,'score':0,'strategy':None} for n in names]

# several rounds
for aRound in range(100):

    # en each round:
    for player1,player2 in itertools.combinations(society,2):
        # each chooses strategy
        player1['strategy']=choice(strategies)

        player2['strategy']=choice(strategies)

        # result from strategy chosen
        result=payoff[player1['strategy'],player2['strategy']]

        # update scores
        player1['score']+=result[0]
        player2['score']+=result[1]

## RESULTADOS POR RONDA
        print(f"{player1['name']} ({player1['strategy']}) vs {player2['name']} ({player2['strategy']}): "
              f"Puntaje - {player1['name']}: {player1['score']}, {player2['name']}: {player2['score']}")
```

```python
print("\nResultados finales:")
for player in society:
    print(f"{player['name']}: {player['score']}")

winner = max(society, key=lambda x: x['score'])
print(f"\nEl gran ganador es: {winner['name']} with a score of {winner['score']}")
```

3. El resultado del gran ganador del torneo
